# Bollinger Bands and how to use them?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
df = yf.download('SPY', '2020-01-01', '2020-12-31')
```

```python
df['Adj Close'].head()
```

# Rolling Mean 

```python
df['SMA_20d'] = df['Adj Close'].rolling(window=20).mean()
df.head(20)
```

```python
# Upper bands
df['Upper Bands'] = df['Adj Close'].rolling(20).mean() + df['Adj Close'].rolling(20).std()*2

#Lower bands
df['Lower Bands'] = df['Adj Close'].rolling(20).mean() - df['Adj Close'].rolling(20).std()*2
```

```python
plt.figure(figsize=[15,10])
plt.grid(True)
plt.plot(df['Adj Close'], label='Adj Close')
plt.plot(df['Upper Bands'], label='Upper bollinger band')
plt.plot(df['Lower Bands'], label='Lower bollinger band')
plt.xlabel('Date')
plt.ylabel('Prices')
plt.legend(loc=2)
plt.show()
```
