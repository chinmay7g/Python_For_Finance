# First stock code on new laptop

```python
# importing the libraries

import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

data = yf.download('GS', '2019-01-01', '2020-01-01')
data.head()
```

```python
plt.figure(figsize=[14,6])
data['Adj Close'].plot()
```

```python
# making a global function

def stock_read(stocks, start, end):
    data = yf.download(stocks, start, end)
    return data['Adj Close'].head()
```

```python
stocks = ['GS', 'JPM']
start = '2019-01-01'
end = '2020-01-01'

stock_read(stocks, start, end)
```

```python
# including class method

class Stock_Analysis:
    
    def stock_read(self, stocks, start, end):
        data = yf.download(stocks, start, end)
        df= data['Adj Close'].head(5)
        return df
    
    def stock_plot(self, data):
        return data.plot()
    
```

```python
stocks = ['GS', 'JPM']
start = '2019-01-01'
end = '2020-01-01'
st = Stock_Analysis()
df = st.stock_read(stocks, start, end)
st.stock_plot(df)
```

```python
start = '2019-01-01'
end = '2020-01-01'
stocks = ['JPM', 'GS', 'BAC', 'BRK-A']

df = stock_read(stocks, start, end)
df.plot()
```
